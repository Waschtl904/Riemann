```python
# Zelle 2: Setup – Importe
import math
import matplotlib.pyplot as plt

from prime_connection import zeta_series, zeta_euler, generate_primes, prime_error
from utils import plot_prime_distribution, plot_euler_vs_series
```

## Zelle 3: Definition der Euler-Produktformel vs. Reihe  
Wir betrachten \(s = \sigma + i\,b\).  
- **zeta_series**: Summen-Darstellung \(\sum_{n=1}^{n_{\max}}n^{-s}\)  
- **zeta_euler**: Produkt-Darstellung \(\prod_{p\le p_{\max}}(1-p^{-s})^{-1}\)


```python
# Zelle 4: Numerischer Vergleich Serie vs. Euler-Produkt
s = complex(2, 10)  # Beispiel s = 2 + 10i
n_max = 500
p_max = 500
series_val = zeta_series(s, n_max)
euler_val = zeta_euler(s, generate_primes(p_max))
print(f"ζ_series({s}) = {series_val:.6f}")
print(f"ζ_euler ({s}) = {euler_val:.6f}")
print(f"Abweichung  = {abs(series_val - euler_val):.2e}")
```

## Zelle 5: Plot – Serie vs. Produkt über Imaginary-Anteile  
Wir variieren den Imaginärteil \(b\) und plotten \(|ζ_{\rm series}-ζ_{\rm euler}|\).


```python
# Zelle 6: Plot Euler vs. Series
plot_euler_vs_series(s_real=2, imag_list=list(range(0, 61, 10)), p_max=500, n_max=500)
```

## Zelle 7: Primzahldichte vs. Näherung  
Wir vergleichen die tatsächliche Anzahl π(x) mit der Näherung x/log(x).


```python
# Zelle 8: Plot Primzahldichte
plot_prime_distribution(1000)
```

## Zelle 9: Fehleranalyse der Primzahldichte  
Wir berechnen den Unterschied π(x) – x/log(x) für ausgewählte x-Werte.


```python
# Zelle 10: Fehleranalyse
for x in [10, 100, 1000, 5000]:
    primes = generate_primes(x)
    err = prime_error(x, primes)
    print(f"x = {x:4d}  π(x) – x/log(x) = {err:.2f}")
```

- Die Abweichungen der Primzahldichte π(x) um x/log(x) lassen sich numerisch analysieren.
